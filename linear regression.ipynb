##small example

```python
##my equation 2x+6
import numpy as np
import pandas as pd
```

```python
X = np.array([8, 7, 2, 15, 40]).reshape(-1, 1)
X_1 = np.c_[np.ones((len(X), 1)), X]
X_1
```

```python
y = 2 * X + 6
```

```python
inverse_of_dot = np.linalg.inv(X_1.T.dot(X_1))
result_W = inverse_of_dot.dot(X_1.T).dot(y)
result_W
```

```python
print(f"{round(result_W[1, 0])}x + {round(result_W[0, 0])}")
```

##On big data

```python
df = pd.read_csv('/content/drive/MyDrive/Iris.csv')
df
```

```python
for i in range(len(df)):
  if(df['Species'][i] == 'Iris-setosa' ):
    df.at[i, "class"] = -1
  else:
    df.at[i, "class"] = 1
df
```

```python
X = df.drop(['Species', 'class','Id'], axis=1)
Y = df[['class']].copy()

X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

Y_train.reset_index(inplace=True, drop=True)
Y_test.reset_index(inplace=True, drop=True)
X_train.reset_index(inplace=True, drop=True)
X_test.reset_index(inplace=True, drop=True)
```

```python
for i in range(len(X_train)):
    if Y_train['class'][i] == -1:
        X_train.loc[i] = X_train.loc[i] * -1
```

```python
X_train_arr = np.array(X_train)
Y_train_arr = np.array(Y_train)
Y_train_arr
```

```python
X_train_arr = np.c_[np.ones((len(X_train_arr), 1)), X_train_arr]
X_train_arr
```

```python
inverse_of_dot = np.linalg.inv(X_train_arr.T.dot(X_train_arr))
result_W = inverse_of_dot.dot(X_train_arr.T).dot(Y_train_arr)
result_W
```

```python
X_test_arr = np.array(X_test)
X_test_arr = np.c_[np.ones((len(X_test_arr), 1)), X_test_arr]
# Y_test_arr = np.array(Y_test)

# X_test_arr

for i in range(len(X_test)):
  check = X_test_arr[i].dot(result_W)
  if(check > 0):
    Y_test.at[i, "check"] = 1
  else:
    Y_test.at[i, "check"] = -1


Y_test
```

accuracy

```python
acc = 0
for i in range(len(Y_test)):
  if(Y_test['check'][i] == Y_test['class'][i]):
    acc+=1
acc = acc/len(Y_test) * 100
print(f"The accuracy = {acc}%")
```
